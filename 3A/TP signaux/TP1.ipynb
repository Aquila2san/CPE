```python
# Import des fonctions
import numpy as np 
import msicpe.san as san 
from plotly import express as px 
import scipy.signal as signal
```

```python
# Question 1.1
# Dfinition des fonctions
t = np.linspace(0, 4 * np.pi, 800)
x = np.cos(3 * t) + np.sin(5 * t)
y = np.cos(np.sqrt(2) * t) + np.sin(3 * t)
z = np.cos(t**2 / 2)
px.line( x = t, y = x, labels={"x":"Temps", "y":"Amplitude"}, title="cos(3t)+sin(5t)").show()
px.line( x = t, y = y, labels={"x":"Temps", "y":"Amplitude"}, title="coV2t)+sin(3t)").show()
px.line( x = t, y = z, labels={"x":"Temps", "y":"Amplitude"}, title="cos(t2/2)").show()
```

```python
# Question 1.2
# Affichage de x(t) et y(t)
t = np.linspace(0, 1, 500)
def x(t):
    return np.cos(2 * np.pi * t)
def y(t):
    return np.sin(2 * np.pi * t)
px.line( x = t, y = x(t), labels={"x":"Temps", "y":"Amplitude"}, title="cos(2pit)").show()
px.line( x = t, y = y(t), labels={"x":"Temps", "y":"Amplitude"}, title="sin(2pit)").show()
```

```python
# Calcul et affichage de la fonction d'inter corrlation
gamma_xy, tau = san.correlate(x(t), y(t), 1)
px.line( x = tau, y = gamma_xy, labels={"x":"Tau", "y":"Amplitude"}, title="gamma_xy").show()
print ("On peut voir que gamma_xy est impaire")
gamma_xy2 = -(1-np.abs(tau))*np.sin(2*np.pi*tau)/2
px.line( x = tau, y = gamma_xy2, labels={"x":"Tau", "y":"Amplitude"}, title="gamma_xy2").show()
```

```python
# Question 1.2.2
import pandas as pd

print ("La fonction gamma_xy est nulle pour tau = 1, tau = 0.5, tau = 0, tau = -0.5 et  tau = -1")

#GnrationduDataFrame 
df_signaux = pd.DataFrame({
    "Temps(s)": np.tile(t, 7),
    "Valeur": np.concatenate([
        x(t),
        y(t),
        x(t) * y(t),
        y(t - 0.5),
        x(t) * y(t - 0.5),
        y(t - 1),
        x(t) * y(t - 1)]),
    "Signal": (["x(t)"] * len(t)
               + ["y(t)"] * len(t)
               + ["x(t) * y(t)"] * len(t)
               + ["y(t - 0.5)"] * len(t)
               + ["x(t) * y(t - 0.5)"] * len(t)
               + ["y(t - 1)"] * len(t)
               + ["x(t) * y(t - 1)"] * len(t))})

# Affichage des signaux
px.line(df_signaux,x="Temps(s)",y="Valeur",color="Signal",title="Signaux").show()

print ("Le domaine d'intgration sur lequel l'intgration a lieu est [tau, 1].")
```

```python
# Question 1.2.3
indice_gamma_max = np.argmax(gamma_xy)
tau_max = tau[indice_gamma_max]
print ("La valeur tau_max de tau pour laquelle gamma_xy atteint son maximum est tau_max =", tau_max)

#GnrationduDataFrame 
df_signaux = pd.DataFrame({
    "Temps(s)": np.tile(t, 3),
    "Valeur": np.concatenate([
        x(t),
        y(t - tau_max),
        x(t) * y(t - tau_max),]),
    "Signal": (["x(t)"] * len(t)
               + ["y(t - tau_max)"] * len(t)
               + ["x(t) * y(t - tau_max)"] * len(t))})

# Affichage des signaux
px.line(df_signaux,x="Temps(s)",y="Valeur",color="Signal",title="tau = tau_max").show()

print ("Le domaine d'intgration est [tau_max, 1]")
```

```python
# Partie 2

# Dfinition des variables 
A = 1 # Amplitude
nu0 = 7 # Frquence de l'impulsion
T = 5 * (1 / nu0) # Dure de l'impulsion
D = 3 * T # Dure de l'observation
t = np.linspace(0, D, 1000)

# Dfinition de x(t)
def porte(t):
    y = (np.abs(t) < 0.5)
    return np.float16(y)
    
def x(t):
    return A * np.cos(2 * np.pi * nu0 * t) * porte((t - T / 2) / T)

# Affichage de x(t)
px.line( x = t, y = x(t), labels={"x":"t", "y":"Amplitude"}, title="x(t)").show()
```

```python
# Question 2.1.1.2
tau0 = 1

def y(t):
    return x(t - tau0)

#GnrationduDataFrame 
df_signaux = pd.DataFrame({
    "Temps(s)": np.tile(t, 2),
    "Valeur": np.concatenate([
        x(t),
        y(t)]),
    "Signal": (["x(t)"] * len(t)
               + ["y(t)"] * len(t))})

# Affichage des signaux
px.line(df_signaux,x="Temps(s)",y="Valeur",color="Signal",title="tau = tau0").show()
```
